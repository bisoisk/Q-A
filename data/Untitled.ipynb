```python
import os

PATH='text_data/S08_set1_a'
PATH2='.txt.clean'
file_range=range(1,10)
labels=[]
txt_data=[]
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
for i in file_range:
    File_Path=PATH+str(i)+PATH2
    print(File_Path)
    file=open(File_Path,'r')
    txt=file.readlines()[1:]    
    txt_data.append(txt)
    topic=txt[0]
    labels.append(topic[:-1])    
    
    
    
```

```python
def remove_stopwords(txt_data):
    from nltk.tokenize import RegexpTokenizer
    tokenizer = RegexpTokenizer(r'\w+')
    data_list=[]
    for data in txt_data:
        paragraph_list=[]
        for paragraph in data:
            word_tokens=tokenizer.tokenize(paragraph)
            filtered_sentence = [w for w in word_tokens if not w in stop_words]
            filtered_sentence = [w for w in filtered_sentence if w]
            #print(filtered_sentence)
            paragraph_list.append(filtered_sentence)
        data_list.append(paragraph_list)
    return data_list    
```

```python
fil_txt_data=remove_stopwords(txt_data)
```

```python
flat_list = [item for sublist in fil_txt_data for item in sublist]
```
